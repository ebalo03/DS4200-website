```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 1(a): times series plot & splitting data

```python
# loading data
data = pd.read_csv('KoreaCovid.csv')
data.head()
```

```python
# extract columns
confirmed = data['Confirmed']
recovered = data['Recovered']
deaths = data['Deaths']

# Plot the time series data
plt.figure(figsize=(14, 7))
plt.plot(confirmed, label='Confirmed Cases')
plt.plot(recovered, label='Recovered Cases')
plt.plot(deaths, label='Deaths')
plt.xlabel('Time (Number of Days Since Janurary 22, 2020)')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases in South Korea')
plt.legend()
plt.show()
```
